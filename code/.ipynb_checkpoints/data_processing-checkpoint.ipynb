```python
import pandas as pd
import numpy as np
datayear=2017
```

```python
bbbsDF = pd.read_csv('../data/Ball_by_Ball.csv')
bbbsDF = bbbsDF.convert_objects(convert_numeric=True)
matchesDF = pd.read_csv('../data/Match.csv')
playerMatchesDF = pd.read_csv('../data/Player_Match.csv')
playersDF = pd.read_csv('../data/Player.csv')
seasonsDF = pd.read_csv('../data/Season.csv')
teamsDF = pd.read_csv('../data/Team.csv')
```

```python
def getMatchesForTeam(id):
    return matchesDF[(matchesDF["Team_Name_Id"] == id) | (matchesDF["Opponent_Team_Id"] == id)]
```

```python
def generatePlayerMatchesCompleteData():
    print("Generating PlayerMatchesComplete Data")
    data = pd.DataFrame(columns=('Player_Id','Player_Name','No_of_Matches','Bat','Bowl','is_bowler'))
    pos = 0
    totalSize = len(playersDF)
    for index,row in playersDF.iterrows():
        print("Processing data...( %d %% done)" % int(pos*100/totalSize) )
        bowl =0
        bat = 0
        playerId = row['Player_Id']
        playerName= row['Player_Name']
        playerAsStrikerDF = bbbsDF[bbbsDF["Striker"] == playerId]
        bat=len(playerAsStrikerDF)
        score=sum(list(playerAsStrikerDF.loc[:,'Runs_Scored']))
        playerAsStrikerDF1 = playerAsStrikerDF[(bbbsDF["Innings_No"])==1 ]
        bat1=len(playerAsStrikerDF1)
        score1=sum(list(playerAsStrikerDF1.loc[:,'Runs_Scored']))
        playerAsStrikerDF2 = playerAsStrikerDF[(bbbsDF["Innings_No"])==2 ]
        bat2=len(playerAsStrikerDF2)
        score2=sum(list(playerAsStrikerDF2.loc[:,'Runs_Scored']))
        playerAsBowlerDF = bbbsDF[bbbsDF["Bowler"] == playerId]
        bowl=len(playerAsBowlerDF)
        Match_played=bbbsDF[(bbbsDF["Bowler"] == playerId) | (bbbsDF["Striker"] == playerId) |( bbbsDF["Non_Striker"] == playerId) | (bbbsDF["Fielders"] == playerId)]
        No_of_Matches=len(set(list(Match_played.loc[:,'MatcH_id'])))
            
        season=list(playerAsStrikerDF.loc[:,'Season'])+list(playerAsBowlerDF.loc[:,'Season'])
        team=list(playerAsStrikerDF.loc[:,'Team_Batting'])+list(playerAsBowlerDF.loc[:,'Team_Bowling'])
        
        teams=[0]*10
        for i in range(len(season)):
            teams[season[i]-2008]= 0 if np.isnan(team[i]) else int(team[i])
        #print(teams)
        
        if(No_of_Matches):
            data.loc[pos,"Player_Id"] = playerId
            data.loc[pos,"Player_Name"] = playerName
            data.loc[pos,"No_of_Matches"] =No_of_Matches
            data.loc[pos,"Bowl"] = bowl
            data.loc[pos,"T_Score"] =score
            data.loc[pos,"Bat"] = bat
            data.loc[pos,"T_Score1"] =score1
            data.loc[pos,"Bat1"] =bat1
            data.loc[pos,"T_Score2"] =score2
            data.loc[pos,"Bat2"] =bat2
            data.loc[pos,"is_bowler"] =1 if bowl/No_of_Matches>=6 else 0
            data.loc[pos,'MPlayed_'+str(2007)]=0
            for i in range(len(teams)):
                data.loc[pos,'Team_'+str(i+2008)]=teams[i]
                i+=2008
                match=Match_played[Match_played["Season"]==i]
                matchp1=Match_played[Match_played["Season"]==i-1]
                matchp2=Match_played[Match_played["Season"]==i-2]
                No_of_Matches=len(set(list(match.loc[:,'MatcH_id'])))+ len(set(list(matchp1.loc[:,'MatcH_id'])))+ len(set(list(matchp2.loc[:,'MatcH_id'])))
                data.loc[pos,"MPlayed_"+str(i)]=No_of_Matches
            pos=pos+1
    data.to_csv("../data/processed/player_temp.csv",sep=',')
    print("Done")
    
generatePlayerMatchesCompleteData();
```

```python
playersnewDF = pd.read_csv('../data/processed/player.csv')
bowlers=(playersnewDF)
bowlers=list(map(int,list(bowlers.loc[:,'Player_Id'])))
print(bowlers)
```

```python
def generatebowlerwise():
    print("Generating Bowler wise Data")
    data = pd.DataFrame()
    pos = 0
    totalSize = len(playersnewDF)
    for index,row in playersnewDF.iterrows():
        print("Processing data...( %d %% done)" % int(pos*100/totalSize) )
        bowl =0
        bat = 0
        playerId = row['Player_Id']
        data.loc[pos,"Player_Id"] = playerId
        for i in bowlers:
            
            playerAsStrikerDF = bbbsDF[(bbbsDF["Bowler"] == i) ] 
            playerAsStrikerDF = playerAsStrikerDF[(playerAsStrikerDF["Striker"] == playerId) ] 
            bat=len(playerAsStrikerDF)
            score=sum(list(playerAsStrikerDF.loc[:,'Runs_Scored']))
            data.loc[pos,"B"+str(i)+"runs"] = score
            data.loc[pos,"B"+str(i)+"balls"] =bat
        pos=pos+1
    data.to_csv("../data/processed/bowlerwise.csv",sep=',')
    print("Done")
    
```
