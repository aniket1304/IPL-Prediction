```python
import pandas as pd
import numpy as np

matchesDF = pd.read_csv('../data/Match.csv')
venueDF = pd.read_csv('../data/venue.csv')
teamsDF = pd.read_csv('../data/Team.csv')
testmatchesDF=pd.read_csv('../data/test_data.csv')
testmatches2019DF=pd.read_csv('../data/test_2019.csv')
```

```python
def getTeamId(name):
    return (list((teamsDF[(teamsDF["Team_Name"] == name)]).loc[:,'Team_Id'])[0]) if len(list((teamsDF[(teamsDF["Team_Name"] == name)]).loc[:,'Team_Id'])) else 0

def getVenueId(name):
    return (list((venueDF[(venueDF["City_Name"] == name)]).loc[:,'vid'])[0]) if len(list((venueDF[(venueDF["City_Name"] == name)]).loc[:,'vid'])) else 0

def getTossDecision(name):
    if (name=="bat"):
        return 0
    else:
        return 1
    
def getWinType(name):
    if (name=="runs"):
        return 0
    else:
        return 1
    
```

```python
def Cleaning_Data():
    print("Cleaning Data")
    pos = 0
    data = pd.DataFrame()
    totalSize = len(matchesDF)
    for index,row in matchesDF.iterrows():
        print("Processing data...( %d %% done)" % int(pos*100/totalSize) )
        Team1 = row['Team1']
        match_id = row['match_id']
        Season_Year = row['Season_Year']
        Team2 = row['Team2']
        city = row['City_Name']
        Toss_Winner = row['Toss_Winner']
        winner = row['match_winner']
        Toss_Name = row['Toss_Name']
        Win_Type = row['Win_Type']
        
        Team1=getTeamId(Team1)
        Team2=getTeamId(Team2)
        city=getVenueId(city)
        winner=getTeamId(winner)
        Toss_Winner = getTeamId(Toss_Winner)
        Toss_Name=getTossDecision(Toss_Name)
        Win_Type=getWinType(Win_Type)
        
        data.loc[pos,"match_id"] = match_id
        data.loc[pos,"Season_Year"] = Season_Year
        data.loc[pos,"Team1"] = Team1
        data.loc[pos,"Team2"] = Team2
        data.loc[pos,"City_Name"] = city
        data.loc[pos,"Toss_Winner"] = Toss_Winner
        data.loc[pos,"Toss_Name"] = Toss_Name
        data.loc[pos,"Win_Type"] = Win_Type
        data.loc[pos,"match_winner"] = winner
        pos=pos+1
    #data.to_csv("../data/processed/match_cleaned.csv",sep=',')
    print('Done')
        
#Cleaning_Data()
```
